```python
import pandas as pd
import seaborn as sb

data = pd.read_csv("../csv/sueldos.csv",sep=",",engine="python")
data['TOTAL SALARIO BRUTO'] = [float(i[1:].replace(',', '.')) for i in data['TOTAL SALARIO BRUTO']]
data['ASIGNACIÓN POR CARGO'] = [float(i[1:].replace(',', '.')) for i in data['ASIGNACIÓN POR CARGO']]

data1 = data[["TOTAL SALARIO BRUTO","AÑO", "APELLIDO_Y_NOMBRE"]]
grafico = sb.pairplot(data1,hue="APELLIDO_Y_NOMBRE",palette="icefire")
```

Podemos concluir que quienes tengan picos más altos en el 1er grafico son quienes tuvieron una mayor cantidad de ingresos brutos, tambien podemos ver que la mayoría de los ingresos son del año 2020
