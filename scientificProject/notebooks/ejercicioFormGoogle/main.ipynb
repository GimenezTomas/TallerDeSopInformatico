```python
import pandas as pd
import matplotlib.pyplot as gf1
import matplotlib.pyplot as gf2
import matplotlib.pyplot as gf3

data = pd.read_csv("../csv/TP4 Análisis de datos.csv")
```

En esta sección se puede apreciar a que se dedican los jóvenes según su edad.

```python
x = data['¿Estudian o trabajan?'].unique()
y = data['¿Estudian o trabajan?'].loc[data['Edad']=='17 a 19'].value_counts().tolist()

gf1.bar(x,y)
gf1.title('Ocupación_menores_de_20')
gf1.xlabel('Ocupación')
gf1.ylabel('Cantidad de individuos')
gf1.xticks(rotation=90,fontsize=12)
gf1.show()
```

Aqui se observa cual es la carrera mas estudiada en universidades privadas

```python
x1 = data['¿Qué estudian?'].loc[data['¿Asisten a una universidad pública o privada? ']=='Privada'].unique()
y1 = data['¿Qué estudian?'].loc[data['¿Asisten a una universidad pública o privada? ']=='Privada'].value_counts().tolist()

gf2.bar(x1,y1)
gf2.title('Carrera más estudiada en universidades privadas')
gf2.xlabel('Carrera')
gf2.ylabel('Cantidad de individuos')
gf2.xticks(rotation=90,fontsize=12)
gf2.show()
```

Por último elegí mostrar la carrera donde se presenta mayor cantidad de estudiantes trabajando

```python

x2 = data['¿Qué estudian?'].loc[data['¿Estudian o trabajan?']=='Estudio y trabajo'].value_counts().index.tolist()
y2 = data['¿Qué estudian?'].loc[data['¿Estudian o trabajan?']=='Estudio y trabajo'].value_counts().tolist()

gf3.bar(x2,y2)
gf3.title('Carrera con más estudiantes trabajando mientras estudian')
gf3.xlabel('Carrera')
gf3.ylabel('Cantidad de individuos')
gf3.xticks(rotation=90,fontsize=12)
gf3.show()
```
